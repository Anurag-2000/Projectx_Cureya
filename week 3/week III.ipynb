## Feature Engineering

```python
final_df = df
final_df.drop('Country', axis = 1,inplace = True)
final_df.drop('Happiness Rank', axis = 1,inplace = True)
final_df.drop('Happiness Score', axis = 1,inplace = True)
final_df = pd.get_dummies(final_df,drop_first = True)
```

```python
final_df.head()
```

## Splitting The Data set

```python
X = final_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Random forest genrator

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 50, random_state = 42 )
regressor.fit(X_train, np.ravel(y_train,order='C'))
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
# print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = regressor, X = X_train, y = np.ravel(y_train,order='C'), cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
